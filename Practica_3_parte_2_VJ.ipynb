```python
def calcular_rms_promedio(epocas):
    """
    epocas: arreglo numpy con forma (n_epocas, n_canales, n_muestras)

    Retorna:
    - rms_por_epoca: matriz (n_epocas, n_canales)
    - rms_promedio: vector (n_canales,)
    """
    # RMS por cada época y canal, calculado sobre el eje del tiempo
    rms_por_epoca = np.sqrt(np.mean(epocas**2, axis=2))

    # Promedio de RMS a través de todas las épocas
    rms_promedio = np.mean(rms_por_epoca, axis=0)

    return rms_por_epoca, rms_promedio
```

```python
carpeta = "/content/PRACTICA3/PRACTICA3"
sujetos = range(1, 11)
runs_validos = [4, 8, 12]

def calcular_rms_promedio(epocas):
    rms_por_epoca = np.sqrt(np.mean(epocas**2, axis=2))   # (n_epocas, n_canales)
    rms_promedio = np.mean(rms_por_epoca, axis=0)         # (n_canales,)
    return rms_por_epoca, rms_promedio

datos_T1 = []
datos_T2 = []
ids_sujetos = []
nombres_canales = None

for sujeto in sujetos:
    epocas_T1_sujeto = []
    epocas_T2_sujeto = []

    for run in runs_validos:
        archivo = os.path.join(carpeta, f"S{sujeto:03d}R{run:02d}.edf")

        if not os.path.exists(archivo):
            print(f"No se encontró: {archivo}")
            continue

        raw = mne.io.read_raw_edf(archivo, preload=True, verbose=False)

        eventos, event_id = mne.events_from_annotations(raw, verbose=False)

        epochs = mne.Epochs(
            raw,
            eventos,
            event_id=event_id,
            tmin=0,
            tmax=4,
            baseline=None,
            preload=True,
            verbose=False
        )

        if nombres_canales is None:
            nombres_canales = epochs.ch_names

        if 'T1' in event_id:
            epocas_T1_sujeto.append(epochs['T1'].get_data())

        if 'T2' in event_id:
            epocas_T2_sujeto.append(epochs['T2'].get_data())

    if len(epocas_T1_sujeto) > 0 and len(epocas_T2_sujeto) > 0:
        epocas_T1_sujeto = np.concatenate(epocas_T1_sujeto, axis=0)
        epocas_T2_sujeto = np.concatenate(epocas_T2_sujeto, axis=0)

        _, rms_T1 = calcular_rms_promedio(epocas_T1_sujeto)
        _, rms_T2 = calcular_rms_promedio(epocas_T2_sujeto)

        datos_T1.append(rms_T1)
        datos_T2.append(rms_T2)
        ids_sujetos.append(f"Sujeto_{sujeto:03d}")

df_T1 = pd.DataFrame(datos_T1, index=ids_sujetos, columns=nombres_canales)
df_T2 = pd.DataFrame(datos_T2, index=ids_sujetos, columns=nombres_canales)

display(df_T1)
display(df_T2)
```

```python
from scipy.stats import shapiro
import pandas as pd

resultados_normalidad = []
normalT1=[]
normalT2=[]

for canal in df_T1.columns:
    stat_T1, p_T1 = shapiro(df_T1[canal])
    stat_T2, p_T2 = shapiro(df_T2[canal])

    resultados_normalidad.append({
        "Canal": canal,
        "Shapiro_T1_stat": stat_T1,
        "Shapiro_T1_p": p_T1,
        "Normal_T1": "Sí" if p_T1 > 0.05 else "No",
        "Shapiro_T2_stat": stat_T2,
        "Shapiro_T2_p": p_T2,
        "Normal_T2": "Sí" if p_T2 > 0.05 else "No"
    })
    if p_T1 > 0.05:
      normalT1.append(canal)
    if p_T2 > 0.05:
      normalT2.append(canal)

df_normalidad = pd.DataFrame(resultados_normalidad)
display(df_normalidad)
print(normalT1)
print(normalT2)
```

```python
from scipy.stats import levene

resultados_levene = []

for canal in df_T1.columns:
    stat, p = levene(df_T1[canal], df_T2[canal])

    resultados_levene.append({
        "Canal": canal,
        "Levene_stat": stat,
        "Levene_p": p,
        "Varianzas_iguales": "Sí" if p > 0.05 else "No"
    })

    if p < 0.05:
      print(canal)

df_levene = pd.DataFrame(resultados_levene)
display(df_levene)

```

# RESULTADOS DE LAS PRUEBAS Y APLICACIÓN DE PRUEBAS PARAMÉTRICAS O NO PARAMÉTRICAS
Se realizaron pruebas de normalidad y homocedaticidad para las tareas 1 y 2 por canal. Antes de analizar los resultados, se debe conocer el parámetro que determina la selección de una hipótesis:

--------------------------------------------------------------------------------
- PRUEBA SHAPIRO WILK:
  
  sirve para determinar si las muestran siguen una distribución normal o no:
  - Ho= los datos provienen de una distribución normal.
  - Ha= los datos no provienen de una distribución normal.

  Para elegir cuál hipótesis se debe aceptar y cuál rechazar se tiene en cuenta un valor estadístico p-valor que se compara con el valor de significancia α (0.05 = 5% de posible error)
  - p≤0.05 --> se rechaza Ho, por lo que los datos no siguen una distribución normal
  - p>0.05 --> se acepta Ho, los datos siguen una distribución normal.
--------------------------------------------------------------------------------
- LEVENE TEST:
  
  Determina si 2 o más grupos tienen la misma varianza (supuesto de homogeneidad de varianzas)

  - Ho: Las varianzas de los grupos son iguales (se acepta cuando p> 0.05)
  - Ha= Las varianzas son diferentes entre sí (se acepta cuando p≤ 0.05)

--------------------------------------------------------------------------------
- PRUEBA T STUDENT:

  La prueba t de Student es una prueba estadística que analiza si dos muestras proceden de poblaciones que cuentan con la misma media. Cuantifica la diferencia entre la media de las dos muestras. (Las diferencias entre los grupos no son producto del azar).

  - Ho: No hay diferencias entre los ciclos elegidos, es decir que las medidas poblacionales son iguales. (no se distingue entre la izquierda y la derecha)

  - Hipótesis alternativa: Sí existe una diferencia entre las medidas de los ciclos. (si se diferencia entre la izquierda y la derecha)

  Las condiciones de un t de Student son:
  - Independencia: Observaciones independientes entre ellas.
  - Normalidad: Las muestras se tienen que distribuir de forma normal.
  - Igualdad de varianza (homocedasticidad): La varianza de ambas poblaciones debe ser igual.

  Se va a asumir un valor para establecer la estadística del problema:
  - p > 0.05: No hay evidencia significativa de diferencia (acepta Ho)
  - p <= 0.05: Hay diferencia significativa (rechaza Ho- canal relevante para distinguir intensiones motoras)

--------------------------------------------------------------------------------
- PRUEBA DE MANN WHITNEY (U TEST):
  
  Busca determinar si 2 grupos provienen de la misma distribución, es decir, si hay o no diferencias reales. Se aplica cuando no siguen una distribución normal.
  - Ho: las distribuciones de ambos grupos son iguales- el canal no diferencia entre la mano izquierda y derecha
  - Ha: las distribuciones son diferentes por lo que el canal sí permite distinguir la tarea entre ambas manos.
  Para determinar si el canal es de utilidad o no se toma de referencia el p-value:
  - p<=0.05: rechaza Ho, quiere decir que hay diferencia significativa
  - p> 0.05: acepta Ho, no hay diferencia significativa y no es un canal relevante.

--------------------------------------------------------------------------------
- p value:

  Teniendo en cuenta la varianza de las mauestras, estima la probabilidad de que se obtenga una diferencia igual o mayor a la que se observa si se establece para aceptar o rechazar un hipótesis nula. Esa probabilidad se conoce como p-value. Cuando p-value es mayor que cierto límite determinado, se dice que no se rechaza la hipótesis nula; si ese valor es menor, se considera que la hipótesis nula se rechaza.

## Resultados:
- En las pruebas de Shapiro wilk para normalidad de los datos, se obtiene que los 14 canales de la tarea 1 (Fc2, C1, Cz, C2, Cp1, Cp2, Cp4, Fz, P1, Pz, P2, P4, Po4, Iz) siguen una distribución normal. Cuando se analizan los canales para la tarea 2 que siguen una distribución normal se obtiene que los canales C1, Cz, C2, Cp1, Cp2, Cp4, Fz, P1, Pz, P2, P4, Po4 también siguen una distribución normal.
- Para la prueba de Levene, la relacipon entre el canal T1 y T2 en todos los casos cumple con el supuesto de homocedasticidad.
- La prueba de T student será aplicada a los canales C1, Cz, C2, Cp1, Cp2, Cp4, Fz, P1, Pz, P2, P4, Po4 ya que cumplen con los supuestos para aplicarse esta prueba. Los demás canales utilizarán la prueba Mann whitney.

```python
resultados_t = []
df_canalt= df_T1[['C1..', 'Cz..', 'C2..', 'Cp1.', 'Cp2.', 'Cp4.', 'Fz..', 'P1..', 'Pz..', 'P2..', 'P4..', 'Po4.']]
from scipy.stats import ttest_ind

for canal in df_canalt:
    stat, p = ttest_ind(df_T1[canal], df_T2[canal])

    resultados_t.append({
        "Canal": canal,
        "ttest_stat": stat,
        "ttest_p": p,
        "Canal relevante": "Sí" if p <= 0.05 else "No"
    })

    if p <= 0.05:
      print(canal)

df_tstudent = pd.DataFrame(resultados_t)
display(df_tstudent)

```

```python
# PRUEBA DE MANN WHITNEY
df_mann = df_T1.drop(columns=['C1..', 'Cz..', 'C2..', 'Cp1.', 'Cp2.', 'Cp4.', 'Fz..', 'P1..', 'Pz..', 'P2..', 'P4..', 'Po4.'])
```

```python
from scipy.stats import mannwhitneyu

resultados_mannwhitney = []

for canal in df_mann.columns:
    stat, p = mannwhitneyu(df_T1[canal], df_T2[canal], alternative='two-sided')

    resultados_mannwhitney.append({
        "Canal": canal,
        "MannWhitneyU_stat": stat,
        "MannWhitneyU_p": p,
        "Canal relevante": "Sí" if p <= 0.05 else "No"
    })
    if p <= 0.05:
      print(canal)

df_mannwhitney = pd.DataFrame(resultados_mannwhitney)
display(df_mannwhitney)
```

# Resultados y análisis:

La imaginería motora activa la corteza motora sun generar movimiento real y produce cambios en el EEG, estos cambios son específicos en ciertas frecuencias (3-12Hz y 13-30Hz) y son de muy baja amplitud. No implica un aumento en la amplitud sino en la redistribución de la energía en el espectro de frecuencias.

La distribución de los datos en la prueba de normalidad (Shapiro Wilk) no son homogéneos entre los canales, es esperable ya que las señales de encefalografía no son homogénas por la variabilidad fisiológica entre sujetos y la naturaleza de la señal EEG no es estacionaria.

De las pruebas t student y Mann Whitney se encontró que no hay diferencias estadísticamente significativas que permita reconocer canales significativos para imaginería de la mano izquierda (T1) y la mano derecha (T2) en ninguno de los canales analizados.

Una de las razones del porqué no se obtuvo un canal significativo puede ir en que, al utilizar una medida como el RMS, se mide la energía y amplitud total y no en una frecuencia específica.

En un artículo visto, la imaginería puede no salir bien al depender mucho del sujeto y no todos los sujetos ejecutan igual la tarea asignada, esto genera una variabilidad alta y por ende una respuesta EEG inconsistente entre sujetos, reduciendo las probabilidad de enconttrar diferencias significativamente significativas en la población en general.

Aunque no se identificaron canales relevantes que permitan identificar las tareas 1 y 2, los resultados indican la necesidad de utilizar métodos más específicos para el análisis del espectro y poder realizar una caracterización correcta de la actividad cerebral en imaginería motora.
