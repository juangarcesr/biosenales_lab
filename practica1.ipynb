```python
#se importan las librerías necesarias para la práctica
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python

#se crea un par de vectores a y b
a = np.array([3.1, 1, -0.5, -3.2, 6])
b = np.array([1, 3, 2.2, 5.1, 1])
```

```python
#se hace multiplicación escalar, es decir, se multiplica componente a componente y se suman los valores.
producto = np.dot(a, b)
print(producto)
#para poder multiplicar los dos vectores deben ser de la misma dimensión.
#la función np.dot sirve para calcular el producto punto.
```

```python
#para realizar la multiplicación punto a punto de los dos vectores basta en poner su multiplicación, esto porque los dos vectores que usamos en este caso tienen la misma dimensión
producto = a * b
print(producto)
```

```python
#se crea la matriz pedida mediante la función np.array
A = np.array([
    [2,   -1,  -3],
    [4,   1.5, -2.5],
    [7.3, -0.9, 0.2]
])

print(A)
```

```python
#round:Redondea los números al entero más cercano o a cierta cantidad de decimales
x = np.array([2.345, 4.567, 7.891])
print(np.round(x))      # entero
print(np.round(x,2))    # 2 decimales
```

```python

# vector n desde 0 hasta 100, se pone 101 porque python no toma el último valor.
n = np.arange(0,101)

# función y[n]
#no se imprime para no tener tantos valores en consola
y = np.sin(np.pi * 0.12 * n)
```

```python
#se asume que es el mismo vector n por lo que no creamos uno de nuevo
# señal y2[n]
y2 = np.cos(2*np.pi*0.03*n)
```

```python

# suma
s = y + y2

# producto punto a punto
t = y * y2
```

```python
# el figsize es como lo dice su nombre para darle el tamaño a la figura que vamos a crear
# gráfica
plt.figure(figsize=(10,5))
#se crean leyendas con colores diferentes para mostrar a qué línea se refiere a cada función
plt.plot(n, y, color='blue', label='y[n] = sin(pi*0.12n)')
plt.plot(n, y2, color='red', label='y2[n] = cos(2pi*0.03n)')
#se crean los nombres de los ejes y también el nombre del gráfico
plt.xlabel('n (muestras)')
plt.ylabel('Amplitud')
plt.title('Señales y[n] y y2[n]')
#se activa la legenta y además se pone la cuadrícula
plt.legend()
plt.grid(True)
#este plt.show no siempre es necesario pero no sobra
plt.show()
```

```python

# gráfica
plt.figure(figsize=(10,5))

plt.plot(n, s, color='green', label='s[n] = y[n] + y2[n]')
plt.plot(n, t, color='purple', label='t[n] = y[n] * y2[n]')

plt.xlabel('n (muestras)')
plt.ylabel('Amplitud')
plt.title('Señales suma y producto')

plt.legend()
plt.grid(True)

plt.show()
```

```python

def estadisticas_notas(dic_notas):
    # convertir a Serie
    notas = pd.Series(dic_notas)
    
    # calcular estadísticas
    resultado = pd.Series({
        'Minima': notas.min(),
        'Maxima': notas.max(),
        'Media': notas.mean(),
        'Desviacion_tipica': notas.std()
    })
    
    return resultado
```

```python


def agregar_BMI(df):
    # convertir altura de cm a metros
    altura_m = df['Height'] / 100
    
    # crear nueva columna BMI
    df['BMI'] = df['Weight'] / (altura_m ** 2)
    
    return df
```

```python
#se lee el archivo csv, como está en la misma carpera del notebook no se debe poner una ruta, solamente el nombre seguido de .csv
#con la función head se imprimen los primeros valores del dataframe

df = pd.read_csv("datos.csv", sep=";")
print(df.head())
```

```python
#se elimina con la función drop la columna unnamed
df = df.drop(columns=["Unnamed: 0"])
```

```python
#se imrpimen nuevamente las primeras filas para ver si efectivamente se eliminó la columna indeseada
print(df.head())
```

```python
# se aplica la función para calcular el imc y adicionar al df.
df = agregar_BMI(df)

print(df.head())
```

```python
#adicional se categoriza según el imc si está en bajo peso, normal, sobrepeso u obesidad.
def clasificar_bmi(df):
    
    def categoria(valor):
        if valor < 18.5:
            return 'Bajo peso'
        elif valor < 25:
            return 'Normal'
        elif valor < 30:
            return 'Sobrepeso'
        else:
            return 'Obesidad'
    
    df['Categoria_IMC'] = df['BMI'].apply(categoria)
    return df
```

```python
df = clasificar_bmi(df)
print(df[['Weight','Height','BMI','Categoria_IMC']].head())
```
