### 1. Descripción del proyecto:
La técnica BCI (Brain-Computer-Interface) consiste en un sistema que codifica señales de la actividad cerebral en tiempo real para interactuar con el entorno sin necesidad de un movimiento real, que enfocado en la rehabilitación motora, la técnica de BCI se utiliza principalmente a reconocer y trasmitir la intención motora del paciente de mover una extremidad que perdió la movilidad a señales que producen una retroalimentación sensorial por medio de dispositivos como robots o estimulación eléctrica.

El BCI tiene 2 enfoques principales que consisten un un enfoque asistivo y un enfoque rehabilitadora. En el primer enfoque, se busca sustituir las vías neuronales dañadas ofreciendo una comunicación constante en el control de dispositivos; el enfoque rehabilitadora busca recuperar las conexiones neuronales que se dañaron y restaurar funciones. 

Para lograr esto, la BCI se apoya directamente con la electroencefalografía (EEG) que puede ser invasiva o no invasiva.para la base de datos utilizada como referencia en el proyecto, se realiza un EEG no invasivo con una distribución 10-10 que permite una cantidad mayor de electrodos ubicados en el cuero cabelludo y, estos sensores van a capturar la actividad eléctrica de diferentes regiones del cerebro. Cuando se le pide al paciente realozar una actividad motora como lo es el movimiento de una mano o pie, aunque no logre realizarlo, el cerebro va a generar patrones que se relacionan a la tarea de movimiento asignada con un área del cerebro, específicamente a la corteza motora (canales C, según estudios previos, los más significativos son C3 y C4); los patrones reflejan cambios en las ondas cerebrales (alpha y beta).

### 3. Historia, uso y proyectos relacionados con BCI
El desarrollo de las interfaces cerebro–computador (BCI) tiene sus ra ́ıces en los primeros estudios sobre la actividad electrica cerebral, iniciados con el electroencefalograma (EEG). Este fue registrado por primera vez por Hans Berger en 1924, permitiendo identificar las ondas alfa y beta, lo que marcó un antes y un después en la comprensión del funcionamiento cerebral. En 1930, Berger publicó su trabajo “Uber das Elektrenkephalogramm des Menschen”, donde presentó oficialmente sus hallazgos [1]. 

El descubrimiento del EEG fue un paso previo fundamental para el desarrollo de los sistemas BCI. No obstante, sus antecedentes se remontan al siglo XIX, cuando Richard Caton registro señales eléctricas en cerebros de animales y publicó sus resultados en 1875 en el British Medical Journal. Paralelamente, los cient ́ıficos polacos Napoleon Nikodem Cybulski y Adolf Beck realizaron experimentos similares, y en 1913, Vladimir Pravdicz-Neminski describio distintos tipos de actividad bioelectrica cerebral, a la que llamó “electrocerebrograma”  ́[1]. 

Estas investigaciones fueron las bases para que, decadas después, los científicos imaginaran la posibilidad de comunicar directamente el cerebro con una maquina. En 1973, el investigador Jacques J. Vidal, de la Universidad de California, adoptó por primera vez el termino Brain–Computer Interface (BCI) y propuso usar senales EEG para controlar computadoras y prótesis, marcando el inicio formal de esta tecnología [2]. 

A finales de los anos 80, Farwell y Donchin (1988) desarrollaron el primer sistema funcional, el “P300 Speller”, que permitía escribir letras mediante los potenciales evocados del EEG. En la decada de 1990, Jonathan Wolpaw y su equipo en EE.UU. demostraron que era posible mover un cursor controlando la amplitud de las ondas, mientras que el grupo de Gert Pfurtscheller en Austria introdujo el prototipo de imaginacion motora, en el que el usuario imagina movimientos para generar patrones cerebrales espec ́ıficos. Desde entonces, las BCI han evolucionado rapidamente gracias al avance de la electronica, la neurociencia y la inteligencia artificial. A  principios del siglo XXI surgieron paradigmas basados en estímulos visuales (SSVEP), y mas recientemente, sistemas apoyados en aprendizaje profundo (Deep Learning), que mejoraron notablemente la precision en la decodificación de señales cerebrales [2].


Actualmente, grandes compañías y centros de investigación están desarrollando BCIs invasivas y no invasivas para restaurar funciones motoras, permitir comunicación a personas paralizadas o controlar protesis robóticas. Aunque aún existen desafíos tecnológicos y éticos, los avances recientes confirman que la electroencefalografía fue el punto de partida esencial para convertir en realidad la comunicacion directa entre el cerebro y las maquinas [2]. 

Algunos de los principales usos que se le confieren al BCI en avances significativos en los ultimos tiempos, se pueden evidenciar que uno de los principales usos esta relacionado con las tecnicas invasivas para decodificación de información motora con el uso de microelectrodos intracorticales con gran resolucion espacial y temporal, como se observa en la *Fig01* [2]. Otro uso que se ha evidenciado en articuo cient ́ıficos es la comunicación del lenguaje por medio de BCI invasivo, en donde se buscaba obtener una comunicación más fluída con la decodificacion de palabras maginadas por medio de microelectrodos instracorticales. Por otro lado, la tecnología BCI no-invasiva basada en EEG con e objetivo de identificar los distintos tipos de movimientos por medio de la detección no invasiva en miembros superiores [2].

Ademas de las aplicaciones ya mencionadas, las interfaces cerebro-computadora (BCIs) tambien están comenzando a explorarse en el ámbito del entretenimiento. Actualmente existen dispositivos comerciales, como cascos con electrodos para uso doméstico, que permiten controlar juegos clásicos como Tetris mediante senales cerebrales. Asimismo, se espera que en los próximos años surjan avances significativos en experiencias inmersivas de realidad virtual basadas en BCIs [2]. Estas aplicaciones no se limitan al ocio. El uso de BCIs en Serious Games ha demostrado beneficios terapeuticos, especialmente para pacientes con trastornos neurologicos como epilepsia o déficit de atención.

Los proyectos futuros estan relacionados con problemáticas identificadas a lo largo de resultados obtenidos en la experimentación actual. Se observó que existen riesgos quirúrgicos relacionados con dispositivos invasivos que requieren una craneotomía para dispositivos de ancho de banda (alta frecuencia de muestreo) que requieren altos niveles de potencia los cuales pueden afectar la vida util de la batería del dispositivo [3]. Otra de las problematicas identificadas corresponde a la imposibilidad del movimiento de la cabeza ya que mucho de los dispositivos cuentan con cableados lo cual se traduce en riesgos de infeccion por diversas reacciones del cuerpo relacionado con respuestas inmunes y el ruido asociado. 

Una de las alternativas probables a futuro es el uso de tecnologías no invasivas como el EEG debido a su alta frecuencia de muestreo, o el fNIRS (Luz IR para medición de oxigenacion cerebral) que cuenta con una mejor resolución espacial. Aunque el sistema no invasivo sea la mejor opción para tratamientos seguros, estas limitaciones se pueden ver expresadas en una baja precision y respuesta [3]. 

Dentro de propuestas futuras se ha evaluado el BCI implantable por medio de venas (Stentrode) en donde se evita el uso de cirugía abierta. Este procedimiento se basa en el uso de electrodos introducidos en los vasos sanguíneos del cerebro por medio de un stent con electrodos. Con esta propuesta se obtuvo resultados de personas con ELA que escribían por medio de clicks mentales adicionandole un seguimiento ocular [3]. 

A futuro se espera que el avance en el BCI se vea reflejado en el uso de electrodos mas densos y miniaturizados, BCI inalámbricas para evitar claves que comprometan la salud del paciente y el resultado final de la senal. Asimismo, el uso de implantes como por ejemplo el Stentrode [3].

### 4. Uso de coherencia y conectividad cerebral en BCI

Las interfaces cerebro-computador (BCI) basadas en electroencefalografía (EEG) han evolucionado significativamente en los últimos años, pasando de enfoques centrados en características locales de la señal a metodologías que consideran la interacción entre distintas regiones del cerebro. En este contexto, la **coherencia y la conectividad cerebral** se han consolidado como herramientas fundamentales para el estudio de la actividad neuronal y la mejora del rendimiento de los sistemas BCI.

De acuerdo con el artículo Recent applications of EEG-based brain-computer-interface (1), los sistemas BCI tienen como objetivo decodificar señales cerebrales para traducirlas en comandos que permitan la interacción con dispositivos externos. Sin embargo, debido a la naturaleza compleja, no estacionaria y de baja amplitud del EEG, los métodos tradicionales basados únicamente en características espectrales presentan limitaciones en términos de precisión.

En este sentido, la **conectividad funcional** surge como una alternativa más representativa de la dinámica cerebral, ya que permite analizar la sincronización y la interacción entre diferentes regiones del cerebro. Esta se define como el estudio de las relaciones estadísticas entre señales neuronales, particularmente en términos de sincronización de fase o correlación temporal (2). Dentro de estas métricas, la **coherencia** es ampliamente utilizada, ya que cuantifica el grado de sincronización entre dos señales EEG en el dominio de la frecuencia, reflejando la comunicación funcional entre regiones cerebrales.

El uso de conectividad en BCI permite capturar la naturaleza distribuida del procesamiento cerebral. En lugar de analizar canales de EEG de forma independiente, se construyen redes donde los nodos representan regiones cerebrales y las conexiones representan la interacción entre ellas. Este enfoque ha demostrado ser especialmente útil para tareas como la imaginación motora, el reconocimiento de estados cognitivos y la clasificación de señales neuronales, ya que proporciona información más significativa sobre la organización funcional del cerebro.

Además, estudios recientes han demostrado que las características basadas en conectividad, como el **Phase Lag Index (PLI)**, la **correlación de envolvente de amplitud (AEC)** y otras métricas de sincronización, mejoran significativamente el rendimiento de los algoritmos de clasificación en comparación con métodos tradicionales [2]. Estas métricas permiten identificar patrones dinámicos de interacción cerebral que cambian en función del estado cognitivo o la tarea realizada.

Por otra parte, Se ha observado que niveles más altos de conectividad están asociados con estados de mayor actividad cognitiva y conciencia, mientras que una reducción en la conectividad refleja alteraciones en la comunicación neuronal (3).
