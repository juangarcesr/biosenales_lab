### Impulso unitario
Ser√° 1 cuando n = n0 sin importar el valor que tome n0

$$
\delta(n-n_0) =
\begin{cases}
1, & \text{si } n = n_0 \\
0, & \text{si } n \neq n_0
\end{cases}
$$


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

def impseq(n0, n1, n2):
# Genera x(n) = delta(n-n0); n1 <= n <= n2
# ----------------------------------------------
 n = np.arange(n1,n2+1) # Se crea el vector de muestras
 x = (n-n0) == 0 
 return [x,n]
```

### Escal√≥n unitario:
Ser√° 1 cuando n sea mayor o igual que n0
$$
u(n-n_0) =
\begin{cases}
1, & \text{si } n \geq n_0 \\
0, & \text{si } n < n_0
\end{cases}
$$

Implemente una funci√≥n en Python que permita generar una funci√≥n escal√≥n unitario, definida en un intervalo ùëõ1‚â§ ùëõ0‚â§ ùëõ2





```python
def unitario(n0, n1, n2):
# ----------------------------------------------
 n = np.arange(n1,n2+1) # Se crea el vector de muestras
 x = (n - n0) >= 0             # condicion para que sea 1 o True
 x = x.astype(int)             # convertir True/False -> 1/0
 return [x,n]
```

### Rampa:
Ser√° 0 cuando ùëõ sea menor que ùëõ0.
Por ejemplo, si
ùëõ = ‚àí2 y ùëõ0 = 0 entonces la rampa es 0.

Si ùëõ‚â•ùëõ0, la rampa toma el valor de (ùëõ‚àíùëõ0)
Por ejemplo, si
ùëõ=4 y ùëõ0=0, entonces la rampa vale 4‚àí0=4.


$$
r(n-n_0) =
\begin{cases}
0, & \text{si } n < n_0 \\
n - n_0, & \text{si } n \geq n_0
\end{cases}
$$

Implemente una funci√≥n en Python que permita generar una funci√≥n rampa, definida en un intervalo ùëõ1 ‚â§ ùëõ0 ‚â§ ùëõ2 con una pendiente m.

```python
import numpy as np
def rampa(n0, n1, n2):
# ----------------------------------------------
 n = np.arange(n1,n2+1) # Se crea el vector de muestras
 x = (n - n0) * (n >= n0)      # condicion para que sea 1 o True
 return [x,n]
```

### Exponenciales complejas

$$
x(n) = e^{(\sigma + j\omega_0)n}, \quad \forall  n
$$

$$
\text{Si } \sigma \text{ es: }
\begin{cases}
> 0, & \text{la se√±al se amplifica} \\
< 0, & \text{la se√±al se aten√∫a}
\end{cases}
$$

Cuando se quiere implementar esta funcion en python se hace uso del comando np.exp en donde se le pasa como par√°metros esos valores que se elevaron a euler.

- En una exponencial compleja se esperan como resultado una funcion real y una imaginaria en forma seno y coseno debido a la formula de euler.

$$
e^{(\sigma + j\omega)n} = e^{\sigma n} \cdot e^{j\omega n}
$$

$$
e^{j\omega n} = \cos(\omega n) + j \sin(\omega n),
$$


$$
e^{(\sigma + j\omega)n} = e^{\sigma n} \big( \cos(\omega n) + j \sin(\omega n) \big)
$$

```python
# Escalar real
print(np.exp(1))   # e^1 = 2.718...

# Vector
print(np.exp([0, 1, 2]))   # [1.0, e, e^2]

# N√∫mero complejo
print(np.exp(1 + 1j))   # e^(1+j) = e^1 * e^(j) = 2.718*(cos1 + j sin1)
```

4.\; Genere la siguiente secuencia:

$$
x[n] =
\begin{cases}
r[n], & 0 \leq n \leq 5, \\[6pt]
r[n-5], & 6 \leq n \leq 11, \\[6pt]
r[n-10], & 12 \leq n \leq 17
\end{cases}
$$


Sabiendo que se esta hablando de una secuencias que usa rampas, se hace uso de:
$$
r(n-n_0) =
\begin{cases}
0, & \text{si } n < n_0 \\
n - n_0, & \text{si } n \geq n_0
\end{cases}
$$

Teniendo en cuenta que para valores mayores o iguales a 0 se est√° hablando de una semejanza entre posici√≥n y valor en el vector x. Con n-5 se ve un desplazamiento del vector al igual que con n-10.

```python
# Generar la secuencia en forma de vector con 10 posiciones desde el 0 hasta el 17

#Se le aplica la funcion rampa para cada uno de los tramos
x1, n1 = rampa(0, 0, 5)     #n0 = 0, n1=0, n2=5 --> n1 es desde donde empieza y n2 hasta donde va el vector
x2, n2 = rampa(5, 6, 11)    #n0 = 5, n1 = 6, n2 = 11
x3, n3 = rampa(10, 12, 17)  #n0=10, n1=12, n2=17

# Concatenar todo en un solo vector
x_total = np.concatenate((x1, x2, x3))
n_total = np.concatenate((n1, n2, n3))

print("n_total (Posiciones) =", n_total)
print("x_total (Vector con valores) =", x_total)
```

```python
#la derivada se define como una diferencia hacia valores pasados y conocidos (n-1) 
derivada = np.diff(x_total)
nd = np.arange(0, len(derivada)) #como se realiz√≥ la derivada, se elimin√≥ una posici√≥n
print(derivada)
```

```python
plt.stem(n_total, x_total, linefmt='b-', label='Original')
plt.stem(nd, derivada, linefmt='r-', label='Derivada')
plt.title("Secuencias obtenidas")
plt.xlabel("n")
plt.ylabel("x[n]")
plt.legend()
plt.show()
```
