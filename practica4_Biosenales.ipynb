## Descripción de las etapas del registro

Los biopotenciales se originan en células excitables, como las células musculares y cardíacas, debido a cambios en el potencial de membrana causados por el movimiento de iones. Estos cambios generan potenciales de acción, cuya propagación da lugar a señales eléctricas que pueden ser registradas desde la superficie corporal. En el caso del electrocardiograma (ECG), la señal representa la actividad eléctrica asociada a la despolarización y repolarización del corazón, mientras que en la electromiografía de superficie (EMG) la señal refleja la activación eléctrica de las fibras musculares y de las unidades motoras [1].

Una vez generada, la señal eléctrica se propaga a través de distintos tejidos del cuerpo, como músculo, grasa, piel y otros medios biológicos. Durante este trayecto, la señal puede atenuarse y contaminarse con distintas fuentes de ruido, como interferencia de la red eléctrica, artefactos por movimiento o variaciones en la impedancia de contacto entre la piel y los electrodos. Por esta razón, el registro de biopotenciales requiere etapas de adquisición sensibles y con buen rechazo al ruido [1].

En la práctica realizada, la captación de la señal se llevó a cabo mediante tres electrodos superficiales. Para el registro de ECG, se ubicaron dos electrodos debajo de las clavículas, uno a cada extremo del tronco superior, y un tercer electrodo en la cresta ilíaca, utilizado como referencia o tierra. Esta configuración permitió medir la diferencia de potencial entre dos puntos del cuerpo, usando un tercer punto para estabilizar la medición y disminuir interferencias. Para el registro de EMG, se colocaron dos electrodos a lo largo de un músculo del antebrazo y el electrodo de referencia se ubicó sobre una prominencia ósea del codo. Esta disposición es adecuada para registrar actividad muscular superficial, ya que busca captar la señal del músculo de interés y reducir la captación de actividad de otras zonas [5].

Posteriormente, los electrodos se conectaron al módulo SHIELD-EKG-EMG de Olimex, el cual corresponde a una etapa de acondicionamiento analógico. Este módulo recibe la señal diferencial analógica proveniente del cuerpo y la adapta para su posterior lectura. De manera general, este tipo de sistemas incluye amplificación diferencial, filtrado y adecuación de la señal, ya que los biopotenciales suelen tener amplitudes pequeñas y son susceptibles a interferencias [1]. Gracias a este acondicionamiento, la señal de salida puede ser leída por un sistema de adquisición digital.

La salida del módulo se conectó a un Arduino, encargado de realizar la conversión analógica-digital de la señal. ## Componentes generales de un equipo de adquisición de biopotenciales

A grandes rasgos, un equipo de adquisición de biopotenciales está conformado por los siguientes componentes:

### 1. Electrodos
Son la interfaz entre el cuerpo y el sistema electrónico. Su función es captar la diferencia de potencial presente en la superficie de la piel y permitir la transferencia de la señal biológica hacia el circuito de adquisición [1].

### 2. Etapa de amplificación
Debido a que los biopotenciales suelen ser de baja amplitud, es necesario amplificarlos antes de procesarlos. Para ello, normalmente se emplea un amplificador diferencial o de instrumentación, con alta impedancia de entrada y buena capacidad de rechazo al ruido en modo común [4].

### 3. Etapa de filtrado
Se utiliza para reducir componentes no deseadas de la señal, como la deriva de línea base, el ruido de alta frecuencia y la interferencia de la red eléctrica. Esto mejora la calidad de la señal antes de su digitalización [1].

### 4. Conversor analógico-digital (ADC)
Convierte la señal analógica continua en datos digitales discretos. La calidad de esta etapa depende de factores como la frecuencia de muestreo y la resolución del conversor [1].

### 5. Unidad de procesamiento
Puede ser un microcontrolador, como Arduino, o un sistema computacional. Esta unidad recibe las muestras digitales, las organiza y las envía para visualización, almacenamiento o análisis posterior [3].

### 6. Software de visualización y almacenamiento
Permite observar la señal en el tiempo, guardar los datos en archivos y aplicar técnicas de procesamiento digital para extraer información relevante de las señales registradas [1].

## Problemática asociada al acoplamiento piel-electrodo

El acoplamiento piel-electrodo corresponde a la interfaz física y eléctrica entre el tejido biológico y el sistema de adquisición. Esta etapa no es ideal, ya que la piel, especialmente la capa más externa de la epidermis, presenta una impedancia relativamente alta y variable, lo que dificulta la transferencia estable de la señal biológica hacia el circuito electrónico. Además, la impedancia de contacto no permanece constante: puede cambiar con la preparación de la piel, la presión del electrodo, la presencia de sudor, el secado del gel y el movimiento relativo entre la piel y el electrodo [11].

La principal problemática de esta interfaz es que puede introducir atenuación de la señal, ruido adicional, deriva de línea base y artefactos por movimiento. En particular, varios estudios muestran que los cambios en la impedancia de la interfaz piel-electrodo están estrechamente relacionados con la aparición de artefactos de movimiento en registros de superficie, especialmente en sistemas portátiles o de monitoreo ambulatorio como los de la práctica. Esto significa que una parte importante de las perturbaciones observadas en ECG, EEG o sEMG no proviene del tejido que se desea medir, sino de la propia interfaz de contacto entre el cuerpo y el electrodo[12].

## Estrategias para mejorar el acoplamiento piel-electrodo

Para mejorar esta interfaz se utilizan varias estrategias. En primer lugar, se busca reducir la impedancia de contacto mediante una adecuada preparación de la piel, lo cual puede incluir limpieza, remoción de grasa, ligera abrasión superficial y uso de gel conductor cuando se emplean electrodos húmedos [11], [12]. También es importante usar electrodos en buen estado, colocarlos sobre una zona anatómica apropiada y asegurar una presión y adhesión suficientemente estables durante todo el registro[13].

Otra estrategia consiste en minimizar el movimiento relativo entre la piel, el electrodo y los cables, ya que este desplazamiento modifica la impedancia de contacto y favorece la aparición de artefactos. Para ello se recomienda fijar bien los cables, evitar tracción mecánica sobre los electrodos y procurar que el sujeto permanezca lo más quieto posible durante la adquisición [14]. Desde el punto de vista del sistema electrónico, también se emplean amplificadores con alta impedancia de entrada y alto rechazo en modo común, de manera que las variaciones en la interfaz afecten menos la calidad final de la señal [15].

En aplicaciones modernas también se estudian materiales y diseños alternativos de electrodos, especialmente para sistemas secos o textiles. Sin embargo, estos suelen enfrentar como reto principal una impedancia de contacto más alta y más variable que la de los electrodos húmedos tradicionales, por lo que el diseño mecánico y electrónico del sistema adquiere aún mayor importancia [15].

## ¿Por qué esta es la etapa más crítica en el registro de señales de superficie?

El acoplamiento piel-electrodo se considera la etapa más crítica en el registro de señales de superficie como sEMG, EEG y ECG porque es el primer punto de contacto entre la fuente biológica y el sistema de medición. Si en esta interfaz la señal se distorsiona o se contamina, las etapas posteriores de amplificación, filtrado y digitalización solo procesarán una versión degradada de la señal original. En otras palabras, una mala interfaz no puede compensarse completamente después con procesamiento digital, porque el problema aparece desde el instante mismo en que la señal entra al sistema.

Además, las señales de superficie suelen tener amplitudes bajas y son muy sensibles a pequeñas perturbaciones. Por ello, variaciones aparentemente pequeñas en la impedancia de contacto pueden traducirse en cambios importantes en la morfología, amplitud o estabilidad del registro. Esta situación es particularmente crítica en EEG, donde las amplitudes son muy pequeñas; en ECG, donde los artefactos pueden alterar la línea base o la forma de las ondas; y en sEMG, donde un mal contacto puede modificar la amplitud aparente de la activación muscular[15].

En la práctica realizada, esta etapa fue especialmente importante porque se trabajó con electrodos superficiales conectados a un módulo Olimex y a un Arduino. Por tanto, la calidad del registro dependió en gran medida de una correcta ubicación de los electrodos, una buena adhesión a la piel y una referencia estable. Cualquier cambio en el contacto, en la presión o en el movimiento del sujeto podía reflejarse directamente en la señal adquirida, tanto en el ECG como en el EMG[14].

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import neurokit2 as nk
from ecgdetectors import Detectors
```

```python
df = pd.read_csv("señal_ECG.csv")
fs = 100  # Hz
t = df["tiempo_s"].values
ecg = df["EMG"].values   # esta columna está mal nombrada, pero contiene el ECG
plt.figure(figsize=(12,4))
plt.plot(t, ecg)
plt.title("Señal 1 de ECG adquirida")
plt.xlabel("Tiempo (s)")
plt.ylabel("Valor ADC")
plt.grid(True, alpha=0.3)
plt.show()
```

```python
df = pd.read_csv("señal_ECG_contraria.csv")
fs = 100  # Hz
t1 = df["tiempo_s"].values
ecg1 = df["EMG"].values
plt.figure(figsize=(12,4))
plt.plot(t1, ecg1)
plt.title("Señal 2 de ECG adquirida")
plt.xlabel("Tiempo (s)")
plt.ylabel("Valor ADC")
plt.grid(True, alpha=0.3)
plt.show()
```

En el laboratorio adquirimos dos veces la señal de ECG, entonces lo primero que se hace es graficar las dos para hacer una inspección visual para saber cuál de las dos es más óptima para trabajar en el entorno python, una vez se hace dicha inspección notamos que la señal 1 es mejor porque su línea base es más estable y tiene un patrón más repetitivo por lo que el posterior análisis se va a hacer sobre la señal 1 de ECG.

```python
signals, info = nk.ecg_process(ecg, sampling_rate=fs)
rpeaks_nk = info["ECG_R_Peaks"]
rpeaks_nk
```

```python
detectors = Detectors(fs)

# uno de los algoritmos más usados dentro de esa librería
rpeaks_det = detectors.pan_tompkins_detector(ecg)
rpeaks_det = np.array(rpeaks_det)

rpeaks_det
```

```python
plt.figure(figsize=(14,4))
plt.plot(t, ecg, label="ECG", linewidth=1)

plt.scatter(t[rpeaks_nk], ecg[rpeaks_nk], color="red", label="R peaks NeuroKit", zorder=3)
plt.scatter(t[rpeaks_det], ecg[rpeaks_det], color="green", marker="x", label="R peaks Detectors", zorder=3)

plt.title("Comparación de detección de picos R")
plt.xlabel("Tiempo (s)")
plt.ylabel("Valor ADC")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
print("Cantidad de picos R detectados con NeuroKit:", len(rpeaks_nk))
print("Cantidad de picos R detectados con Detectors:", len(rpeaks_det))
```

Al comparar visualmente la detección de picos R, se observa que el algoritmo de NeuroKitpresenta un mejor desempeño sobre esta señal. Los picos detectados por este método coinciden de manera más consistente con los máximos principales del complejo QRS y muestran una separación temporal más razonable entre latidos consecutivos. Por el contrario, el algoritmo de Detectors identifica una cantidad considerablemente mayor de eventos, muchos de los cuales parecen corresponder a fluctuaciones secundarias de la señal y no a verdaderos picos R.

En términos cuantitativos, NeuroKit detectó 15 picos R en un registro de aproximadamente 10 s, lo que equivale a una frecuencia cardíaca cercana a 90 lpm, valor fisiológicamente plausible para este tipo de práctica. En cambio, Detectors detectó 25 picos R, equivalente a aproximadamente 150 lpm, lo que sugiere una sobredetección asociada a falsos positivos. Por tanto, para esta señal se considera que NeuroKit detecta mejor los picos R.

Sin embargo, estos resultados deben interpretarse con precaución, ya que el registro disponible tiene una duración aproximada de 10 s. Esta longitud es insuficiente para realizar una caracterización robusta de la HRV, especialmente en el dominio de la frecuencia, donde normalmente se requieren registros de mayor duración. Por tanto, en este caso las gráficas obtenidas son útiles principalmente como una exploración ilustrativa del comportamiento de los intervalos R-R, más que como una caracterización clínica o fisiológica concluyente de la variabilidad cardíaca.

```python
import numpy as np

def time_domain_features_manual(nn_intervals):
    nn = np.array(nn_intervals, dtype=float)

    diff_nn = np.diff(nn)
    abs_diff_nn = np.abs(diff_nn)

    features = {
        "mean_nni": np.mean(nn),
        "sdnn": np.std(nn, ddof=1) if len(nn) > 1 else np.nan,
        "sdsd": np.std(diff_nn, ddof=1) if len(diff_nn) > 1 else np.nan,
        "rmssd": np.sqrt(np.mean(diff_nn**2)) if len(diff_nn) > 0 else np.nan,
        "median_nni": np.median(nn),
        "range_nni": np.max(nn) - np.min(nn) if len(nn) > 0 else np.nan,
        "cvsd": (np.sqrt(np.mean(diff_nn**2)) / np.mean(nn)) if len(diff_nn) > 0 and np.mean(nn) != 0 else np.nan,
        "cvnni": (np.std(nn, ddof=1) / np.mean(nn)) if len(nn) > 1 and np.mean(nn) != 0 else np.nan,
        "mean_hr": 60000 / np.mean(nn) if np.mean(nn) != 0 else np.nan,
        "max_hr": 60000 / np.min(nn) if len(nn) > 0 and np.min(nn) != 0 else np.nan,
        "min_hr": 60000 / np.max(nn) if len(nn) > 0 and np.max(nn) != 0 else np.nan,
        "std_hr": np.std(60000 / nn, ddof=1) if len(nn) > 1 else np.nan,
        "nni_50": np.sum(abs_diff_nn > 50),
        "pnni_50": 100 * np.sum(abs_diff_nn > 50) / len(diff_nn) if len(diff_nn) > 0 else np.nan,
        "nni_20": np.sum(abs_diff_nn > 20),
        "pnni_20": 100 * np.sum(abs_diff_nn > 20) / len(diff_nn) if len(diff_nn) > 0 else np.nan,
    }

    return features
```

```python
time_features_nk = time_domain_features_manual(rr_nk)
time_features_det = time_domain_features_manual(rr_det)

features_df = pd.DataFrame([time_features_nk, time_features_det],
                           index=["NeuroKit", "Detectors"]).T
features_df
```

El valor de range_nni también fue mucho mayor con NeuroKit (790 ms) que con Detectors (240 ms), lo cual sugiere que los intervalos detectados por NeuroKit cubren un rango temporal más amplio. Esto puede deberse a la presencia de variaciones reales entre latidos, pero también a la sensibilidad del resultado frente a registros cortos y a posibles errores residuales de detección. De forma consistente con esto, pnni_50 y pnni_20 fueron elevados en ambos métodos, aunque mayores en NeuroKit. Estos índices representan el porcentaje de diferencias sucesivas entre intervalos RR que superan 50 ms o 20 ms, respectivamente, y en este caso sugieren una alta variabilidad latido a latido; no obstante, nuevamente, el corto tiempo de adquisición limita una interpretación fisiológica sólida.

En conjunto, los resultados indican que las características calculadas a partir de NeuroKit son más coherentes con la inspección visual de la señal y con una frecuencia cardíaca fisiológicamente razonable. Por el contrario, las características derivadas de Detectors parecen estar afectadas por la sobredetección de picos R, lo que reduce artificialmente los intervalos RR y eleva la frecuencia cardíaca promedio calculada. Por esta razón, para el análisis posterior de HRV se considera más confiable el conjunto de picos detectado por NeuroKit.

Inicialmente se planteó el uso de la función get_time_domain_features de la librería hrvanalysis para extraer las características de HRV en el dominio del tiempo. Sin embargo, durante la implementación se presentaron problemas de compatibilidad entre dicha librería y algunas dependencias del entorno de Python, lo que impidió su importación y uso correcto.

Debido a ello, se implementó la función time_domain_features_manual(investigada en internet), que calcula manualmente las métricas temporales a partir de los intervalos RR. Esta decisión es válida porque las características del dominio del tiempo no dependen de un algoritmo propietario de la librería, sino de definiciones matemáticas directas sobre la serie de intervalos R-R. Por ejemplo, métricas como mean_nni, sdnn, rmssd, mean_hr, pnni_50 y pnni_20 pueden obtenerse de manera explícita mediante operaciones estadísticas básicas sobre los intervalos detectados.

Por tanto, el uso de time_domain_features_manual no modifica el fundamento del análisis, sino que constituye una implementación equivalente y transparente de las métricas requeridas. Además, esta estrategia permitió mantener el flujo del procesamiento, comparar ambos algoritmos de detección y obtener las características solicitadas aun cuando la librería original no pudo utilizarse en el entorno de trabajo.

## Aplicaciones clínicas del ECG y del EMG

A continuación se presentan tres aplicaciones clínicas del **ECG** y tres del **EMG**, indicando para cada una el tipo de registro más habitual y el procesamiento que suele emplearse para extraer información útil.

### Tabla de aplicaciones clínicas

| Tipo de señal | Aplicación clínica | Tipo de registro usado | Procesamiento típico |
|---|---|---|---|
| **ECG** | **Diagnóstico de arritmias** | ECG de 12 derivaciones en reposo cuando la arritmia está presente durante la consulta, o **ECG ambulatorio** (Holter/parches/event recorders) cuando los episodios son intermitentes [16]. | Filtrado para línea base e interferencia de red, detección de complejos QRS/picos R, cálculo de intervalos RR, análisis de regularidad del ritmo y clasificación morfológica de latidos o episodios arrítmicos [16]. |
| **ECG** | **Detección de isquemia e infarto agudo de miocardio** | Principalmente **ECG de 12 derivaciones**, a menudo serial o complementado con monitorización continua del segmento ST en algunos pacientes [17]. | Filtrado básico, alineación y promediado de latidos cuando aplica, delimitación de QRS-T y análisis del **segmento ST**, onda T y cambios topográficos entre derivaciones [17]. |
| **ECG** | **Evaluación de prolongación del QT y riesgo de torsades de pointes** | ECG de 12 derivaciones, telemetría o monitorización en pacientes expuestos a fármacos que prolongan el QT o con sospecha de síndrome de QT largo [18]. | Delineación del inicio del QRS y final de la onda T, medición del **QT**, corrección por frecuencia cardíaca (**QTc**) y seguimiento seriado para detectar prolongación clínicamente relevante [18]. |
| **EMG** | **Análisis clínico de la marcha y neurorehabilitación** | **sEMG multicanal** de músculos de miembros inferiores, frecuentemente sincronizado con eventos de marcha, video, cinemática o foot-switches [19]. | Filtrado pasabanda, rectificación, suavizado/envolvente lineal, detección de **onset/offset**, análisis temporal de activación muscular y comparación de patrones con referencia normal o pre/post tratamiento [19]. |
| **EMG** | **Diagnóstico de trastornos neuromusculares** | Principalmente **EMG de aguja**, a menudo complementado con estudios de conducción nerviosa, para evaluar actividad espontánea, reclutamiento y potenciales de unidades motoras [20]. | Inspección de actividad en reposo e inserción, análisis de potenciales de unidad motora (amplitud, duración, polifasia), patrones de reclutamiento e interpretación conjunta con datos clínicos y neurofisiológicos [20]. |
| **EMG** | **Evaluación clínica de la deglución y tamizaje de disfagia** | **sEMG** de músculos submentonianos y/o infrahioideos durante maniobras de deglución [21]. | Filtrado, rectificación, cálculo de amplitud y duración de la activación, detección del inicio/fin del evento deglutorio y comparación entre tareas o condiciones clínicas [21]. |


En el caso del ECG, las aplicaciones clínicas más clásicas siguen siendo el diagnóstico de arritmias, la identificación de isquemia o infarto y la evaluación de la repolarización ventricular mediante el intervalo QT. La selección del registro depende del problema clínico: cuando se necesita una visión anatómica y eléctrica global del corazón se utiliza el ECG de 12 derivacione, mientras que para eventos intermitentes es más útil el registro ambulatorio prolongado**. En todos los casos, el procesamiento busca mejorar la calidad de la señal y extraer eventos eléctricos relevantes, como complejos QRS, intervalos RR, cambios del ST o prolongación del QT [18].

En el EMG, la utilidad clínica es más diversa según si se emplea sEMG o EMG de aguja. El sEMG se usa con frecuencia en análisis funcionales, como la marcha o la deglución, porque permite registrar de manera no invasiva el patrón temporal de activación muscular. En cambio, el EMG de aguja se utiliza más en neurología clínica para estudiar directamente la actividad muscular y apoyar el diagnóstico de trastornos neuromusculares [21].

## Conclusiones

1. El proceso de adquisición de biopotenciales permitió evidenciar que la calidad de las señales de superficie depende en gran medida de la correcta ubicación de los electrodos, de la estabilidad del contacto piel-electrodo y del control de fuentes de ruido durante el registro. En la práctica, tanto para ECG como para EMG, se observó que pequeñas variaciones en el montaje o en el movimiento del sujeto pueden afectar notablemente la morfología y estabilidad de la señal adquirida.

2. En el análisis del ECG, la comparación entre los algoritmos de detección de picos R mostró que NeuroKit presentó un desempeño más confiable que Detectors para la señal seleccionada, ya que identificó una cantidad de latidos más coherente con la inspección visual de la señal y con una frecuencia cardíaca fisiológicamente razonable. Esto permitió calcular intervalos R-R y métricas de HRV más consistentes para el registro disponible.

3. El análisis de HRV en el dominio del tiempo mostró que es posible extraer características relevantes a partir de los intervalos R-R detectados; sin embargo, también permitió concluir que la duración del registro influye de forma importante en la confiabilidad de los resultados. Dado que la señal analizada tuvo una duración aproximada de 10 s, las métricas obtenidas deben interpretarse con precaución y entenderse principalmente como una aproximación académica al procesamiento de señales cardíacas, más que como una caracterización clínica robusta.

4. En el caso del EMG, la práctica permitió observar de forma clara la diferencia entre el estado de reposo, la contracción muscular y la relajación posterior. Durante la contracción, la señal presentó un incremento en amplitud y variabilidad, lo cual es consistente con el mayor reclutamiento de unidades motoras. Esto confirma que el EMG superficial es una herramienta útil para estudiar patrones de activación muscular de manera no invasiva.

5. De forma general, el desarrollo de esta práctica permitió integrar conceptos de instrumentación biomédica, adquisición de señales y procesamiento digital, mostrando que el análisis de ECG y EMG no depende únicamente del algoritmo utilizado, sino también de la calidad del registro original. Por ello, una buena adquisición experimental constituye la base para obtener resultados válidos y correctamente interpretables.
